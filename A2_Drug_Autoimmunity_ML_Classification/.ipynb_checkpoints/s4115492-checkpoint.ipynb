# Hypothesis 


#### MolWt vs HeavyAtomCount: 
Hypothesis: Molecular weight might correlate with heavy atom count, havy molecular might have more DIA-positive.


#### TPSA vs NumHAcceptors: 

H-acceptors and bigger polar regions area  might have more DIA-positive.


#### BertzCT vs HeavyAtomCount:

#### RingCount vs NumSaturatedRings:

Ring number or Saturated rings can correlate to DIA-positive.


#### NumRotatableBonds vs MolWt:

Larg molecules can be longer and have more rotatable bonds, and more flexibility might attribute to DIA-positive.


```python
import numpy as np

def calculate_drug_correlation(descriptor1, descriptor2, epsilon=1e-8):
    
    # Average value
    mean_desc1 = np.sum(descriptor1) / (len(descriptor1) + epsilon)
    mean_desc2 = np.sum(descriptor2) / (len(descriptor1) + epsilon)
    
    # compute  corr
    # Subtract the mean (normalizing)
    desc1_sub_mean = descriptor1 - mean_desc1
    desc2_sub_mean = descriptor2 - mean_desc2
    
    #  similarity formula
        # dot product / (magnitude Ã— magnitude) 
        # similarity = dot_product / (length_of_vector_1 Ã— length_of_vector_2)
        # â†’ (+1)best +correlation, (-1)best -correlation, (0)no correlation
        # Compute squared terms and their square roots for the denominator
        # TNormalisation part in correlation. Scales the result so it's always between -1 and 1.
    desc1_centered_sq_sum = np.sum(np.square(desc1_sub_mean))
    desc2_centered_sq_sum = np.sum(np.square(desc2_sub_mean))
    desc1_centered_sq_sum_sqrt = np.sqrt(desc1_centered_sq_sum)
    desc2_centered_sq_sum_sqrt = np.sqrt(desc2_centered_sq_sum)
    
    # Calculate correlation 
    # (standardized covariance between molecular descriptors) Think length = sqrt(xÂ² + yÂ² + ...)
    correlation = np.sum(desc1_sub_mean * desc2_sub_mean) / (desc1_centered_sq_sum_sqrt * desc2_centered_sq_sum_sqrt + epsilon)
    
    return correlation

# Usage example for your pairs:
print('Correlation: \n')
for feature1, feature2 in pair_features:
    corr = calculate_drug_correlation(X_train[feature1], X_train[feature2])
    print( feature1,' and ', feature2 ,' : \ncorr = ', round(corr, 2), '\n')
```

## Task 3: Data Modelling 



```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import numpy as np
```

```python
full_data = pd.concat([train, test], ignore_index=True)
```

```python
correlations = []
for feature in X_train.columns:
    corr = calculate_drug_correlation(X_train[feature], y_train)
    correlations.append((feature, corr))
    

# sort by absolute value
correlations.sort(key = lambda x: abs(x[1]))
# reverse
correlations = correlations[::-1] 
print('TOP 60 strong correlation with DIA label:\n')

feature_list = []
for i, (feature, corr) in enumerate(correlations[:60], 1):
    feature_list.append(feature)
    print('Abs corr: ', round(corr, 3), ' the feature: ', feature)
    
```

```python
X_train = full_data.drop(columns=['SMILES', 'Label'])  # Features only
y_train = full_data['Label']  
X_train = X_train[feature_list]
```

```python
from sklearn.model_selection import KFold 
from sklearn.neighbors import KNeighborsClassifier
KNeighbors_Classifier_score = []

mymodel = KNeighborsClassifier(7)
# Set up KFold
kf = KFold(n_splits=5, shuffle=True, random_state=4)

for k, (train_index, test_index) in enumerate(kf.split(X_train)):
    # print("\n[Train:]", train_index, "\n", "[Test]", test_index)
    
    # Split the data(input)
    X_train_fold = X_train.iloc[train_index]
    X_test_fold = X_train.iloc[test_index]
    # Split the target labels(outcome) 
    y_train_fold = y_train.iloc[train_index]
    y_test_fold = y_train.iloc[test_index]

    # train the model 
    mymodel.fit(X_train_fold, y_train_fold)
    # get the score
    # {0} Will be replaced by the first argument in .format(...), which is k (the fold number)
    # start from sec argument index = 1, format as a float
    print('\nKNeighbors Classifier')
    score =  mymodel.score(X_test_fold, y_test_fold)
    KNeighbors_Classifier_score.append(score)
    print("[fold{0}] score: {1:.3f}".format(k,score))
```

```python
DecisionTree_Classifier_score = []

for k, (train_index, test_index) in enumerate(kf.split(X_train)):

    # DO NOT overwrite X_train or y_train
    X_fold_train = X_train.iloc[train_index]
    X_fold_test = X_train.iloc[test_index]
    y_fold_train = y_train.iloc[train_index]
    y_fold_test = y_train.iloc[test_index]

    # Train and evaluate
    clf = DecisionTreeClassifier(max_depth=5, random_state=42)
    clf.fit(X_fold_train, y_fold_train)
    y_pred = clf.predict(X_fold_test)

    acc = accuracy_score(y_fold_test, y_pred)
    DecisionTree_Classifier_score.append(acc)
    print('\nDecisionTree Classifier')
    print("[fold{0}] score: {1:.3f}".format(k, round(acc, 4)))
```

```python
mean_DecisionTree = np.mean(DecisionTree_Classifier_score)
mean_KNeighbors = np.mean(KNeighbors_Classifier_score)

if mean_DecisionTree > mean_KNeighbors:
    print(" Decision Tree is better than KNN on average.")
else:
    print(" KNN is better than Dec,ision Tree on average.")
```
